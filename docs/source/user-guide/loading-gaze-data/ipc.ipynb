# IPC (Feather) Files

Previously saved gaze data in IPC (Feather) format can be reloaded quickly:

```python
from pymovements.gaze.io import from_ipc

gaze = from_ipc("gaze_data.feather")
```

This is the fastest way to restore already processed datasets. Although, we do not recommend using `feather` files for data that is meant to be shared.
